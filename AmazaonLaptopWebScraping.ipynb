```python

# assignment method 
from selenium import webdriver
import time
import urllib
import matplotlib.pyplot as plt
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

```

```python
options = Options()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
#options.add_argument("--headless")

driver = webdriver.Chrome(options=options)
#driver.delete_all_cookies()
```

```python
i =0
#driver.set_window_size(1024, 768)
driver.get("https://www.amazon.com/s?k=laptops&crid=1ZWGRTOB40YXF&sprefix=%2Caps%2C777&ref=nb_sb_ss_recent_1_0_recent")
elements = driver.find_elements("xpath","//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
product_urls = [element.get_attribute("href") for element in elements]
elements_with_price={}
for product_url in product_urls:
    print(i)
    #element.click()
    driver.get(product_url)
    time.sleep(3)
    #driver.manage().timeouts().implicitlyWait(20, TimeUnit.SECONDS);
    key = driver.find_element(By.ID,"productTitle")
    try:
        price = driver.find_element('xpath',"//span[@class='a-price aok-align-center reinventPricePriceToPayMargin priceToPay']").text
    except:
        price = None  
    try:
        rate =driver.find_element('xpath',"//div[@id='averageCustomerReviews_feature_div']/div[@id='averageCustomerReviews']/span[@class='a-declarative']/span[@id='acrPopover']/span[@class='a-declarative']/a[@role='button']/span[@class='a-size-base a-color-base']").text
    except:
        rate=None
    try:
        rate_num = driver.find_element('xpath',"//span[@id='acrCustomerReviewText']").text
    except:
        rate_num=None
    try:
        image_url = driver.find_element('xpath',"//div[@id='imgTagWrapperId']/img").get_attribute("src")
        urllib.request.urlretrieve(str(image_url),"img{}.jpg".format(i))
    except:
         image_url=None
    
    elements_with_price.setdefault(key.text,[price,rate,rate_num,image_url])
    i=i+1
    
```

```python
i=0
for product,properties in elements_with_price.items():
    print(product)
    print("Price = " + ".".join(str(properties[0]).split()))
    print("Rating = " + str(properties[1]))
    print("Number of Ratings = " + str(properties[2]).split()[0])
    if not (str(properties[3])=="None"):
        print("img url = "+ str(properties[3]))
        img=plt.imread("img{}.jpg".format(i))
        plt.imshow(img)
        plt.show()
    else:
        print("img url = "+ str(properties[3]))
    i+=1
    print()
```
