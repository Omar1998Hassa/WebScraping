```python

# assignment method 
from selenium import webdriver
import time
import urllib
import matplotlib.pyplot as plt
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
```

```python
options = Options()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
#options.add_argument("--headless")

driver = webdriver.Chrome(options=options)
#driver.delete_all_cookies()
```

```python
i =0
#driver.set_window_size(1024, 768)
driver.get("https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW")
movie_title = driver.find_elements("xpath","//td[@class='a-text-left mojo-field-type-title']")
movie_Lifetime_Gross=driver.find_elements("xpath","//td[@class='a-text-right mojo-field-type-money']")
movie_year= driver.find_elements("xpath","//td[@class='a-text-left mojo-field-type-year']")
```

```python
with open("MovieScrap.txt", "w") as my_file:
    for i in range(len(movie_title)):
        
        my_file.write("movie_title : "+ movie_title[i].text+ ", movie_Lifetime_Gross : " + movie_Lifetime_Gross[i].text + " movie_year : "+movie_year[i].text+"\n")
        print("movie_title : "+ movie_title[i].text+ ", movie_Lifetime_Gross : " + movie_Lifetime_Gross[i].text + " movie_year : "+movie_year[i].text+"\n")
```
